### Import pandas, numpy and matplotlib

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

### Import the dataset from this [address]('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv')

````python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
    
df = pd.read_csv(url, sep = '\t')
````

### Show the first 10 entries

````python
df.head(10)
````

### What is the number of observations in the dataset?

````python
len(df)
````

### What is the number of columns in the dataset?

````python
len(df.columns)
````

### Which was the most-ordered item and how many times it was ordered?

````python
df["item_name"].value_counts()[:1] 
````

### How many different items are sold?

````python
len(df.groupby("item_name"))
````

### How many items were orderd in total?

````python
df.groupby("item_name").sum()["quantity"].sum()
````

### Turn the item price into a float

````python
df["item_price_clear"] = df["item_price"].apply(lambda x: float(x[1:]))
````

````python
df.head()
````

### What is the price of each item?

````python
df[["item_name", "item_price"]].drop_duplicates().reset_index()
````

### How much was the revenue for the period in the dataset?

````python
int(round((df["item_price_clear"] * df["quantity"]).sum()))
````

### How many orders were made in the period?

````python
df["order_id"].value_counts().count()
````

### What is the average revenue amount per order?

````python
((df["item_price_clear"] * df["quantity"]).sum()) / (df["order_id"].value_counts().count())
````

### Merge the table ```dfa``` that contains order date and order id with ```df``` and plot daily revenue 

````python
dfa = pd.DataFrame({"date": np.random.choice(pd.date_range(start="20170101", end="20170201"), size=1800),
                   "order_id": np.arange(1800)})
df["cost"] = df["quantity"]*df["item_price_clear"]
dfa["cost"] = df.groupby("order_id").sum()["cost"]
plt.plot(dfa.groupby("date").sum()["cost"], color = "red")
plt.xlabel("Date")
plt.ylabel("Daily revenue")
plt.show()
````

### Create a scatterplot with the number of items orderered per order price

````python
plt.scatter(df.groupby("item_price_clear").sum()["quantity"].reset_index().values[:,0],df.groupby("item_price_clear").sum()["quantity"].reset_index().values[:,1], color = "red")
plt.xlabel("Price")
plt.ylabel("Quantity")
plt.show()
````
