```python
import numpy as np
def main():
    matrix = np.random.random((10,3))
    print(matrix,"\n",np.array([min(matrix[i], key = lambda x: abs(x-0.5)) for i in range(0,10)]).reshape(10,1))
if __name__ == "__main__":
    main()
```

```python
import numpy as np
def main():
    matrix = np.random.random((6,6))
    print(matrix,"\n",np.array([sum(matrix[i])/min(matrix.transpose()[i]) for i in range(0,6)]).reshape(6,1))
if __name__ == "__main__":
    main()
```

```python
import numpy as np
def main():
    x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])
    print(max([x[i] for i in range(1,len(x)) if x[i-1] == 0]))
if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def _kmeans(obs, k):
    centroids = np.random.rand(k,2)
    y_old = []
    while True:
        y = np.array([np.array([np.linalg.norm(obs[i]-centroids[j]) for j in range(k)]).argmin() for i in range(obs.shape[0])]).reshape(obs.shape[0],1)
        if (np.array_equal(y, y_old)):
            break
        centroids = np.array([[np.mean([obs[i][s] for i in range(obs.shape[0]) if y[i] == j] + [centroids[j][s]]) for s in range(2)] for j in range(k)])
        y_old = y
    return [y[:,0], sum([pow(np.linalg.norm(obs[i] - centroids[y[i]]),2) for i in range(obs.shape[0])])]
def main():
    k = int(input("Input number of centroids: "))
    n = int(input("Input number of points that we want to classify: "))
    obs = np.random.rand(n,2)
    [y, distortion] = _kmeans(obs, k)
    print("\nDistortion:", distortion)
    plt.scatter(*obs.T, c = y, s = 10)
    plt.show()
if __name__ == "__main__":
    main()
```
